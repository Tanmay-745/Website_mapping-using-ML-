## Task 2: Text Extraction & Variable Detection

Testing the extraction logic using `mammoth` and identifying variable placeholders using LLM.

```python
import mammoth

def extract_text_from_docx(file_path):
    with open(file_path, "rb") as docx_file:
        result = mammoth.extract_raw_text(docx_file)
        return result.value

template_path = os.path.join(NOTICE_FOLDER, 'Sarvagram_LRN.docx')
extracted_text = extract_text_from_docx(template_path)
print("Sample Extracted Text (First 500 chars):\n", extracted_text[:500])
```
